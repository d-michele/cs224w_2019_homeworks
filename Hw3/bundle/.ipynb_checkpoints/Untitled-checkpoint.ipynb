### Probability of a Path Existing Between Two Randomly Chosen Nodes

When we sample two vertexes from the union of in, out and scc component the sample space is made by all the permutation with repetition which are 9, 4 of which allow us to reach the second vertex starting from the first one: IN -> SCC, IN -> OUT, SCC -> SCC, SCC -> OUT. The probability of drawing each vertex from a particular set is trivially obtained as 
$$p_{Set} = \frac{|Set|}{|In \cup SCC \cup Out|}$$
With this in mind the expect number of the fraction of reachable can be computed as the sum of the probability of each of the 4 favorable cases that can be arranged as:
$$\frac{(|SCC| + |In|)(|SCC| + |Out|)}{(|SCC| + |In| + |Out|)^2}$$

```python
m_disc, m_in, m_out, m_scc, m_tendril_tubes = q1.get_regions_sets(g_mail)
# m_nodes = g_mail.GetNodes()
m_nodes = len(m_in) + len(m_scc) + len(m_out)
path_fraction_email = (len(m_in) + len(m_scc))*(len(m_scc) + len(m_out))/ m_nodes**2
print(f'Expectation of existing path between two nodes on e_mail net: {path_fraction_email}')

e_disc, e_in, e_out, e_scc, e_tendril_tubes = q1.get_regions_sets(g_epinons)
# e_nodes = g_epinons.GetNodes()
path_fraction_epinions = (len(e_in) + len(e_scc))*(len(e_scc) + len(e_out))/ e_nodes**2

print(f'Expectation of existing path between two nodes on epinions net: {path_fraction_epinions}')
```

## Link Analysis
### Personalized PageRank I

Pageranks are obtained by iteratively computing a linear combination of the elements of rank vector r:
$$r_j = A_{j1}r_1 + A_{j2}r_2 + ... + A_{jn}r_n, \; \; \forall i$$
In matrix notation: $$r_{new}=Ar_{old}$$
and in particular:
$$A= \beta M + (1 - \beta) \frac{T}{\vert \mathbb{T}\vert}$$
where $\beta$ is the teleport probability, and T is a matrix with one on all the rows whose index belong to the teleport set $\mathbb{T}$.

<!-- Under the intuition derived from linearity property we have that $f(\alpha_1 x_1 + \alpha_2 x_2 + ... + \alpha_n x_n)= \alpha_1 f(x_1) + \alpha_2 f(x_2) + ... + f(\alpha_n x_n)$. In this particular case however we need to be careful to preserve rank vector scale. -->


From a random walk perspective $\tilde{M}$ the ij-th element indicates with which probability we move from node i to j. Matrix $M$ is identical for the computation of all personalized rank vector, the update rule is 

$$r \leftarrow A r = \beta M + (1-\beta) \frac{T}{|\mathbb{T}|}$$

In computing the different personalized pageranks the only difference in computing the users rank is in the $T$ matrix which is obtained by assigning a uniform probability $1/\mathbb{T}$ over the nodes of interest.
If we denominate $T_A$ the teleport matrix for Agatha and similarly $T_B$ the Bertha one etc. we can express Eloise teleport probability as:

$$
T_E = 3 (\frac{T_A}{3} + \frac{T_C}{3} - \frac{T_B}{3} - \frac{2}{3} T_D)
$$
Or:
$$
T_E = \alpha_A \frac{T_A}{3} + \alpha_C \frac{T_C}{3} - \alpha_B \frac{T_B}{3} - \alpha_D {T_D}
$$
With $\alpha_A = 3$, $\alpha_C = 3$, $\alpha_B=-3$, $\alpha_D= -2$ 

By similarly combining a generic personalized pagerank vector at convergence we get:
$$
\begin{align}
r_N &= \alpha_A r_A + \alpha_C r_C - \alpha_B r_B - \alpha_D r_D = \\
&= \alpha_A A_A r_A + \alpha_B A_B r_B + \alpha_C A_C r_C + \alpha_D A_D r_D  \\ 
&= [\alpha_A \beta M r_A  + \alpha_A(1-\beta) T_A r_A +  \alpha_B \beta M r_B  + \alpha_B (1-\beta) T_B r_B + \alpha_C \beta M r_C  + \alpha_C (1-\beta) T_C r_C + \alpha_D \beta M r_D  + \alpha_D (1-\beta) T_D r_D]  \\
&= \beta M (\alpha_A r_A + \alpha_B r_B +\alpha_C r_C + \alpha_D r_D) + (1-\beta)( \alpha_A T_A r_A + \alpha_B T_B r_B + \alpha_C T_C r_C + \alpha_D T_D r_D) \\
\end{align}
$$

Note that rank vector are stochastic, and as a consequence $\sum_i r_i = 1$. With this in mind $T_N r_N = e_{\mathbb{T}_N}$ which represent a zero vector except for the elements which are in the interest set $\mathbb{T}_N$
$$
\begin{align}
    T_A r_A &= e_{1,2,3} \\
    T_B r_B &= e_{3,4,5} \\
    T_C r_C &= e_{1,4,5} \\
    T_D r_D &= e_{1}
\end{align}
$$

$$
\begin{align}
r_N &=\beta M (\alpha_A r_A + \alpha_B r_B +\alpha_C r_C + \alpha_D r_D) + (1-\beta)( \alpha_A T_A r_A + \alpha_B T_B r_B + \alpha_C T_C r_C + \alpha_D T_D r_D)\\
&=  \beta M (\alpha_A r_A + \alpha_B r_B +\alpha_C r_C + \alpha_D r_D) + (1-\beta)(\alpha_A e_{1,2,3} + \alpha_B e_{3,4,5} + \alpha_C e_{1,4,5} + \alpha_D e_1)
\end{align}
$$


i) In particular for  $\alpha_A = 3$, $\alpha_C = 3$, $\alpha_B=-3$, $\alpha_D= -2$ we have:
$$
\begin{align}
r_E &= \beta M (\alpha_A r_A + \alpha_B r_B +\alpha_C r_C + \alpha_D r_D) + (1-\beta)(\alpha_A e_{1,2,3} + \alpha_B e_{3,4,5} + \alpha_C e_{1,4,5} + \alpha_D e_1)\\
&=\beta M (r_E) + (1-\beta)e_{2} = \beta M (r_E) + (1-\beta)T_E r_E
\end{align}
$$
Note that $\sum_N \alpha_N = 1$ is required in order to have $\sum_i r_E = 1$ 

ii) Unfortunately there is no way to express $T_F$ by combining the other teleport matrices.

iii) $r_G$ can be computed as $r_G = \frac{3}{5}r_A + \frac{3}{10}r_B + \frac{3}{10}r_C - \frac{1}{5}r_D$

### Personalized PageRank II
The set of all the personalized PageRank vectors that can be derived from the already computed set of vectors $V$ are the span of these vectors with alpha coefficients such that they sum up to 1 in order to preserve rank vectors stochasticity.
$$
{\displaystyle \operatorname {span} (V)=\left\{{\left.\sum _{i=1}^{V}\alpha_{i}v_{i} \right| v_{i}\in V, \sum_i \alpha_{i}=1} \right\}.}
$$

### A different equation for PageRank

\begin{align}
r &= Ar \\
&= \beta Mr + \frac{(1-\beta)}{N}\mathbf{1}\mathbf{1}^T r
\end{align}
By definition the matrix multiplication correspond to $\sum_j^N 11_{ij}r_j$ which is equal to 1 for each row $i$, and leads to: 
$$
\beta Mr + \frac{(1-\beta)}{N}\mathbf{1}
$$

### More Hill-Climbing...

With similar approach to 4.2 we can consider a bipartite graph family with $l$ node on the left each of which is connected with $d$ elements  and an $(l+1)th$ node which shares $d/l + 1$ node with each of the other $l$ nodes.
<img src="4-4.jpg" width="500"/>

Then:
$$
\begin{align}
f(S_k) + \sum_{i=1}^k \delta_i - f(T) = 2 f(S_k) - f(T) < b 
\end{align}
$$
obtained with applying the delta definition: $\delta_i = S_{i+1} - S_i$. 
Specifically for the defined family of graphs we have $f(S_1)= d+l+1$, and $f(S_i)=(d - d/l)$ for $1<i \le k-1$, whereas $f(T)=k(d+1)$.

It follows:
$$
\begin{align}
2 f(S_k) - f(T) &= 2((d+l+1) + (k-1)(d - \frac{d}{l}) - k(d+1) \\
&= 2d + 2l + 2 + 2kd - 2k\frac{d}{l} - 2d + 2\frac{d}{l} - kd - k \\ 
&= 2l + 2 + kd - 2k\frac{d}{l} + 2\frac{d}{l} - k
\end{align}
$$
For a fixed $k$ we can act on $l$ and $d$ however if make $l > k$ we can incur in situation which the optimal solution also include the $l+1)th$ node. This motivates the enforcement of $k=l$:
$$
k + 2 + d(k - 2) + 2\frac{d}{k} > b
$$
By acting on $d$ we can exceed any b.
